```python
import os
import pandas as pd
from dotenv import load_dotenv

from langchain_core.documents import Document
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS

```

```python
# --------------------------------------------------
# 0. нҷҳкІҪ ліҖмҲҳ лЎңл“ң
# --------------------------------------------------
load_dotenv()

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")


```

```python
# --------------------------------------------------
# 1. кІҪлЎң м„Өм •
# --------------------------------------------------
RAG_DATA_PATH = "./rag_documents.csv"
PAIR_DATA_PATH = "./response_pairs.csv"

RAG_SAVE_PATH = "./vectorstore/rag"
EXAMPLE_SAVE_PATH = "./vectorstore/example"
```

```python
# --------------------------------------------------
# 2. нҢҢмқј мЎҙмһ¬ м—¬л¶Җ нҷ•мқё
# --------------------------------------------------
def check_file_exists(path):
    if not os.path.exists(path):
        raise FileNotFoundError(
            f"нҢҢмқјмқ„ м°ҫмқ„ мҲҳ м—ҶмҠөлӢҲлӢӨ: {path}\n"
            f"нҳ„мһ¬ мһ‘м—… кІҪлЎң: {os.getcwd()}\n"
            f"нҳ„мһ¬ нҸҙлҚ” нҢҢмқј лӘ©лЎқ: {os.listdir()}"
        )
```

```python
#--------------------------------------------------
# 3. CSV лЎңл“ң
# --------------------------------------------------
def load_data():
    check_file_exists(RAG_DATA_PATH)
    check_file_exists(PAIR_DATA_PATH)

    rag_df = pd.read_csv(RAG_DATA_PATH)
    pair_df = pd.read_csv(PAIR_DATA_PATH)

    print(f"rag_documents.csv shape: {rag_df.shape}")
    print(f"response_pairs.csv shape: {pair_df.shape}")

    return rag_df, pair_df
```

```python
# --------------------------------------------------
# 4. rag_documents.csv -> Document ліҖнҷҳ
#    л¬ёл§Ҙ кІҖмғүмҡ© л¬ём„ң
# --------------------------------------------------
def build_rag_documents(rag_df):
    rag_docs = []

    for _, row in rag_df.iterrows():
        text = f"""
кҙҖкі„: {row['relation']}
мғҒнҷ©: {row['situation']}
нҷ”мһҗ к°җм •: {row['speaker_emotion']}
мІӯмһҗ н–үлҸҷ: {row['listener_behavior']}

м „мІҙ лҢҖнҷ”:
{row['full_dialogue']}
""".strip()

        doc = Document(
            page_content=text,
            metadata={
                "dialogue_id": row["dialogue_id"],
                "source": "rag_documents"
            }
        )
        rag_docs.append(doc)

    return rag_docs
```

```python
# --------------------------------------------------
# 5. response_pairs.csv -> Document ліҖнҷҳ
#    мң мӮ¬ мқ‘лӢө мҳҲмӢң кІҖмғүмҡ© л¬ём„ң
# --------------------------------------------------
def build_example_documents(pair_df):
    example_docs = []

    for _, row in pair_df.iterrows():
        text = f"""
кҙҖкі„: {row['relation']}
мғҒнҷ©: {row['situation']}
нҷ”мһҗ к°җм •: {row['speaker_emotion']}

мқ‘лӢө м§Ғм „ л¬ёл§Ҙ:
{row['context_before_response']}

мІӯмһҗ мқ‘лӢө:
{row['listener_response']}
""".strip()

        doc = Document(
            page_content=text,
            metadata={
                "dialogue_id": row["dialogue_id"],
                "listener_empathy": row.get("listener_empathy", ""),
                "terminate": row.get("terminate", ""),
                "source": "response_pairs"
            }
        )
        example_docs.append(doc)

    return example_docs
```

```python
# --------------------------------------------------
# 6. Document Chunking
# --------------------------------------------------
def split_documents(docs, chunk_size=500, chunk_overlap=100):
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=chunk_size,
        chunk_overlap=chunk_overlap
    )
    split_docs = text_splitter.split_documents(docs)
    return split_docs
```

```python
# --------------------------------------------------
# 7. FAISS Vector Store мғқм„ұ л°Ҹ м ҖмһҘ
# --------------------------------------------------
def build_and_save_vectorstore(docs, save_path, embeddings):
    vectorstore = FAISS.from_documents(docs, embeddings)
    vectorstore.save_local(save_path)
    print(f"м ҖмһҘ мҷ„лЈҢ: {save_path}")
```

```python
def main():
    print("нҳ„мһ¬ мһ‘м—… кІҪлЎң:", os.getcwd())
    print("нҳ„мһ¬ нҸҙлҚ” нҢҢмқј лӘ©лЎқ:", os.listdir())

    # вң… м—¬кё° м•Ҳм—җ мһҲм–ҙм•ј н•Ё
    from langchain_openai import OpenAIEmbeddings

    # вң… API KEY нҸ¬н•Ё (н•ң лІҲл§Ң!)
    embeddings = OpenAIEmbeddings(
        model="text-embedding-3-small",
        api_key="sk-м—¬кё°м—җ_л„Ҳмқҳ_API_KEY"
    )

    # 1) лҚ°мқҙн„° лЎңл“ң
    rag_df, pair_df = load_data()

    # 2) Document мғқм„ұ
    rag_docs = build_rag_documents(rag_df)
    example_docs = build_example_documents(pair_df)

    print(f"мӣҗліё RAG л¬ём„ң мҲҳ: {len(rag_docs)}")
    print(f"мӣҗліё Example л¬ём„ң мҲҳ: {len(example_docs)}")

    # 3) Chunk л¶„н• 
    split_rag_docs = split_documents(rag_docs, chunk_size=500, chunk_overlap=100)
    split_example_docs = split_documents(example_docs, chunk_size=500, chunk_overlap=100)

    print(f"л¶„н• лҗң RAG chunk мҲҳ: {len(split_rag_docs)}")
    print(f"л¶„н• лҗң Example chunk мҲҳ: {len(split_example_docs)}")

    # вқҢ мқҙ мӨ„ мӮӯм ңн•ҙм•ј н•Ё (мӨ‘ліө мғқм„ұ)
    # embeddings = OpenAIEmbeddings(model="text-embedding-3-small")

    # 4) м ҖмһҘ нҸҙлҚ” мғқм„ұ
    os.makedirs("./vectorstore", exist_ok=True)

    # 5) FAISS vector store мғқм„ұ л°Ҹ м ҖмһҘ
    build_and_save_vectorstore(split_rag_docs, RAG_SAVE_PATH, embeddings)
    build_and_save_vectorstore(split_example_docs, EXAMPLE_SAVE_PATH, embeddings)

    print("лӘЁл“  FAISS vector store мғқм„ұмқҙ мҷ„лЈҢлҗҳм—ҲмҠөлӢҲлӢӨ.")
```
